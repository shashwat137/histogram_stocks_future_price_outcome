## The histogram function
  
To predict the future price of a stock, I am going to try to use the following idea from Dividend Discount Model-  
$$ Rate =  \frac{D}{PV} + g$$

Where growth rate (g) is, $$ \frac{EPS-DPS}{EPS} $$  
Also, the future price can be calculated using the following considering compounding -

$$ FV = PV * (1 + Rate\%)^{n/12} $$ 

$$ Future \space return = FV - PV $$

Where n is number of months.

```python
def histogram(**kwargs):
    '''
    Since it is given that only the three parameters are allowed, hence, the dataframe is not being taken as an argument.
    Instead it is loaded everytime the function is called.
    However, as an option, the dataframe can also be used in the function using the argument data.
    Similarly bins is an optional argument meant to set how many bins to set for the histogram as well as the returned data.
    
    kwargs is a dictionary which has the keys- metric, value, timeframe.
    metric is a string. value is a float, timeframe is an integer.
    Please note, metric must be written in the same way as the columns.
    For example, it must contain epsNtm instead of epsntm.
    
    Returns a dataframe of binned returns and their frequencies.
    '''
    
    # Default values
    if 'metric' not in kwargs:
        kwargs['metric'] = 'roe'
    if 'value' not in kwargs:
        kwargs['value'] = 50
    if 'timeframe' not in kwargs:
        kwargs['timeframe'] = 12
    if 'bins' not in kwargs:
        kwargs['bins'] = 10 # Here I have set 10, but we can also use 4 to give the quartiles.
    if 'data' not in kwargs:
        # Load the Data into a DataFrame.
        kwargs['data'] = pd.read_csv('Large_cap_US_Stocks_5yrs.csv')
    
    # Renaming only to keep below code as more readable.
    data = kwargs['data']
    
    # Two types of Metrics are used, ones which must be dervived as a column, and other that must be picked. 
    # They are separated by a forward slash. For ones without a slash, we directly pick the column.
    if '/' in kwargs['metric']:
    
        # Purpose is to separate the metric into the actual two columns that need to be used.
        Metric = kwargs['metric'].split('/')    
        data[kwargs['metric']] = data[Metric[0]] / data[Metric[1]]
        
        # Remove nan values
        data.dropna(axis = 0, inplace=True, subset=[Metric[0], Metric[1]])
    
    else:
        pass
    
    # Add column of Future Price Return.
    data['future_price_ret'] = data['price'] * (((1 + (data['div']/data['price'] + (1 - data['div']/data['epsNtm']))/100)**(kwargs['timeframe']/12)) - 1)
    
    #Remove all NaN values from columns of the dataframe that are being considered.
    data.dropna(axis = 0, inplace=True, subset=['future_price_ret', kwargs['metric']])
    
    
    # Plot the Histogram
    plt.figure(figsize=(10,5))
    plt.title(f'Histogram of Future Price Returns Price for {kwargs["metric"]} greater than {kwargs["value"]} in a timeframe of {kwargs["timeframe"]} months')
    plt.xlabel(f'Future Price Return')
    plt.ylabel(f'Prices Frequency')
    plt.hist(data[data[kwargs['metric']] > kwargs['value']]['future_price_ret'], bins=kwargs['bins'], edgecolor='white')
    plt.show()
    
    # Returning binned returns and frequency
    df = data[data[kwargs['metric']] > kwargs['value']].copy()
    df['ranges'] = pd.cut(df['future_price_ret'], bins = 10)
    df2 = pd.DataFrame(df['ranges'].value_counts())
    df2.reset_index(inplace=True)
    df2 = df2.sort_values(by=['index'])
    df2.reset_index(inplace=True, drop=True)
    df2.rename(columns={"index":"binned_returns", "ranges":"frequency"}, inplace=True)
    
    return df2
```
